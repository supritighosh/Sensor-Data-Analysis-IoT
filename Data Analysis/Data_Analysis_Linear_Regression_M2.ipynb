```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('firebase-1618885447171.csv')
df['minutes'] = df['time'].str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]) + int(x[2])/60)
del df['time']
df.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.20, random_state=0)
#save the data
x_train = train['humidity'].values.reshape(-1,1)
y_train = train['minutes']

x_test = test['humidity'].values.reshape(-1,1)
y_test = test['minutes']
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print(model.intercept_) 
print(model.coef_)
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score

pred = model.predict(x_test) 
mean_squared_error(y_test, pred)
```

```python
from statsmodels.regression.linear_model import OLS
from statsmodels.tools import add_constant

regr = OLS(y_train, add_constant(x_train)).fit()
print(regr.aic)
```

```python
M2 = pd.DataFrame({'actual':y_test, 'predicted':pred})
M2.index = range(1,len(y_test)+1)
M2.plot(figsize=(10,10))
```

```python

plt.figure(figsize=(10,10))
plt.scatter(y_test, pred, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(pred), max(y_test))
p2 = min(min(pred), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('Actual', fontsize=15)
plt.ylabel('Predicted', fontsize=15)
plt.axis('equal')
plt.savefig('m2.png')
plt.show()

```
