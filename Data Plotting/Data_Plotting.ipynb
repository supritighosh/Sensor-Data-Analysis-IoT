#### Import Libraries

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

#### Load Dataset

```python
df = pd.read_csv('firebase-1618885447171.csv')
df['time'] = df['time'].str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]) + int(x[2])/60)

df.head()
```

##### Plot 1: Line plots for temperature data showing 5 lines in each plot corresponding to 5 days of data

```python
sns.set_palette("PuBuGn_d")
plt.figure(figsize=(10,10))
ax = sns.lineplot(x='time', y='temp', hue='day', legend="full", data=df)
ax.set(xlabel="Time", ylabel="Temperature")
plt.savefig('plot1.png')
plt.show()
```

##### Plot 2: Line plots for humidity data showing 5 lines in each plot corresponding to 5 days of data

```python
plt.figure(figsize=(10,10))
ax2 = sns.lineplot(x='time', y='humidity', hue='day', legend="full", data=df)
ax2.set(xlabel="Time", ylabel="Humidity")
plt.savefig('plot2.png')

plt.show()
```

##### Plot 3: Box plots on temperature data where each box corresponds to one days data

```python
plt.figure(figsize=(10,10))
ax3 = sns.boxplot(x='day', y='temp', data=df)
ax3.set(xlabel="Day", ylabel="Temperature")
plt.savefig('plot3.png')

plt.show()
```

##### Plot 4: Box plots on humidity data where each box corresponds to one days data

```python
plt.figure(figsize=(10,10))
ax4 = sns.boxplot(x='day', y='humidity', data=df)
ax4.set(xlabel="Day", ylabel="Humidity")
plt.savefig('plot4.png')

plt.show()
```

##### Plot 5: Line grid plot for any 3 days of data where labels on the are temperature day1, humidity day 1, temperature day 2, humidity day 2,temperature day 3, humidity day 3

```python
from pandas.plotting import scatter_matrix

df_day1= df.loc[df['day'] == 1]
day1_temp = df_day1["temp"]
day1_humidity = df_day1["humidity"]

df_day2= df.loc[df['day'] == 2]
day2_temp = df_day2["temp"]
day2_humidity = df_day2["humidity"]

df_day3= df.loc[df['day'] == 3]
day3_temp = df_day3["temp"]
day3_humidity = df_day3["humidity"]


df_t = pd.DataFrame({'Temp Day1': day1_temp, 'Humd Day1': day1_humidity, 'Temp Day2': day2_temp, 'Humd Day2': day2_humidity, 'Temp Day3': day3_temp, 'Humd Day3': day3_humidity})

plt.savefig('plot5.png')
pd.plotting.scatter_matrix(df_t, diagonal='kde', figsize=(10,10))
```
